# Criando um Scripting <a name="import"></a>


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os


sns.set_theme()  

sinascmar = pd.read_csv('https://raw.githubusercontent.com/Jefersonfranca/EBAC_Curso_Cientista_de_Dados/main/M%C3%B3dulo%2014%20Scripting/input/SINASC_RO_2019_MAR.csv')

sinascabr = pd.read_csv('https://raw.githubusercontent.com/Jefersonfranca/EBAC_Curso_Cientista_de_Dados/main/M%C3%B3dulo%2014%20Scripting/input/SINASC_RO_2019_ABR.csv')

sinascmai = pd.read_csv('https://raw.githubusercontent.com/Jefersonfranca/EBAC_Curso_Cientista_de_Dados/main/M%C3%B3dulo%2014%20Scripting/input/SINASC_RO_2019_MAI.csv')

sinascjun = pd.read_csv('https://raw.githubusercontent.com/Jefersonfranca/EBAC_Curso_Cientista_de_Dados/main/M%C3%B3dulo%2014%20Scripting/input/SINASC_RO_2019_JUN.csv')

sinascdez = pd.read_csv('https://raw.githubusercontent.com/Jefersonfranca/EBAC_Curso_Cientista_de_Dados/main/M%C3%B3dulo%2014%20Scripting/input/SINASC_RO_2019_DEZ.csv')





def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).sort_values(value).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).unstack().plot(figsize=[15, 5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None


#Mar√ßo
plota_pivot_table(sinascmar, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento')

os.makedirs('./imagens/Mar', exist_ok=True)

plota_pivot_table(sinascmar, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./imagens/Mar/media idade mae por sexo.png')

plota_pivot_table(sinascmar, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./imagens/Mar/media peso bebe por sexo.png')

plota_pivot_table(sinascmar, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Mar/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinascmar, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Mar/media apgar1 por gestacao.png')

plota_pivot_table(sinascmar, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
plt.savefig('./imagens/Mar/media apgar5 por gestacao.png')


#Abril
plota_pivot_table(sinascabr, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento')

os.makedirs('./imagens/Abr', exist_ok=True)

plota_pivot_table(sinascabr, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./imagens/Abr/media idade mae por sexo.png')

plota_pivot_table(sinascabr, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./imagens/Abr/media peso bebe por sexo.png')

plota_pivot_table(sinascabr, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Abr/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinascabr, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Abr/media apgar1 por gestacao.png')

plota_pivot_table(sinascabr, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
plt.savefig('./imagens/Abr/media apgar5 por gestacao.png')


#Maio
plota_pivot_table(sinascmai, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento')

os.makedirs('./imagens/Mai', exist_ok=True)

plota_pivot_table(sinascmai, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./imagens/Mai/media idade mae por sexo.png')

plota_pivot_table(sinascmai, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./imagens/Mai/media peso bebe por sexo.png')

plota_pivot_table(sinascmai, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Mai/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinascmai, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Mai/media apgar1 por gestacao.png')

plota_pivot_table(sinascmai, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
plt.savefig('./imagens/Mai/media apgar5 por gestacao.png')


#Junho
plota_pivot_table(sinascjun, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento')

os.makedirs('./imagens/Jun', exist_ok=True)

plota_pivot_table(sinascjun, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./imagens/Jun/media idade mae por sexo.png')

plota_pivot_table(sinascjun, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./imagens/Jun/media peso bebe por sexo.png')

plota_pivot_table(sinascjun, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Jun/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinascjun, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Jun/media apgar1 por gestacao.png')

plota_pivot_table(sinascjun, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
plt.savefig('./imagens/Jun/media apgar5 por gestacao.png')


#Dezembro
plota_pivot_table(sinascdez, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento')

os.makedirs('./imagens/Dez', exist_ok=True)

plota_pivot_table(sinascdez, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./imagens/Dez/media idade mae por sexo.png')

plota_pivot_table(sinascdez, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./imagens/Dez/media peso bebe por sexo.png')

plota_pivot_table(sinascdez, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/Dez/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinascdez, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./imagens/dez/media apgar1 por gestacao.png')

plota_pivot_table(sinascdez, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
plt.savefig('./imagens/dez/media apgar5 por gestacao.png')
```
