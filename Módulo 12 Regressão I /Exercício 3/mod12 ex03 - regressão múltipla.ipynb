### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf

%matplotlib inline
````

````python
tips = load_dataset(name='tips')

tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']

tips
````

````python
y, X = patsy.dmatrices('tip ~ sex + smoker + time + net_bill', data=tips)
X
````

````python
y
````

````python
# Remova as variáveis não significantes.
y, X = patsy.dmatrices(formula_like='tip ~ net_bill', data=tips)

modelo_sm1 = sm.OLS(y, X).fit()
modelo_sm1.summary()
````

````python
# Observando o gráfico de resíduos em função de net_bill
sns.scatterplot(x='net_bill', y=modelo_sm1.resid, data=tips)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
# Testando transformar net_bill no log e um polinômio. 
y, X = patsy.dmatrices(formula_like='tip ~ np.log(net_bill)', data=tips)

modelo_sm1_log = sm.OLS(y, X).fit()
modelo_sm1_log.summary()
````

````python
y, X = patsy.dmatrices(formula_like='tip ~ net_bill + np.power(net_bill, 2)', 
                       data=tips)

modelo_sm1_pow = sm.OLS(y, X).fit()
modelo_sm1_pow.summary()
````

````python
print('`modelo_sm1_log` R-squared:', round(100*modelo_sm1_log.rsquared,2),"%")
print('`modelo_sm1_pow` R-squared:', round(100*modelo_sm1_pow.rsquared,2),"%")

print("Portanto o modelo_sm1_pow é o melhor modelo por ter um R² maior com:", round(100*modelo_sm1_pow.rsquared,2),"%" )
````

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, diner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
y, X = patsy.dmatrices(formula_like='np.log(tip) ~ sex + smoker + time + net_bill', data=tips)

modelo_sm2 = sm.OLS(y, X).fit()
modelo_sm2.summary()
````

````python
# Removendo as variáveis não signficantes
y, X = patsy.dmatrices(formula_like='np.log(tip) ~ net_bill', data=tips)

modelo_sm2 = sm.OLS(y, X).fit()
modelo_sm2.summary()
````

````python
# Observando o gráfico de resíduos em função de net_bill
sns.scatterplot(x='net_bill', y=modelo_sm2.resid, data=tips)
plt.axhline(y=0, color='r', linestyle='--')

plt.show()
````

````python
# Teste transformar net_bill no log e um polinômio
y, X = patsy.dmatrices(formula_like='np.log(tip) ~ np.log(net_bill)', data=tips)

modelo_sm2_log = sm.OLS(y, X).fit()
modelo_sm2_log.summary()
````

````python
y, X= patsy.dmatrices(formula_like='np.log(tip) ~ net_bill + np.power(net_bill,2)', data=tips)

modelo_sm2_pow = sm.OLS(y, X).fit()
modelo_sm2_pow.summary()
````

````python
print('`modelo_sm2_log` R-squared:', round(100*modelo_sm2_log.rsquared,2),"%")
print('`modelo_sm2_pow` R-squared:', round(100*modelo_sm2_pow.rsquared,2),"%")

print("Portanto o modelo_sm2_pow é o melhor modelo por ter um R² maior com:", round(100*modelo_sm2_pow.rsquared,2),"%" )
````

````python
# Calculando o  R² na escala de tip (sem o log) e comparando com o modelo do item 1


tips['pred_tip_log'] = np.exp(modelo_sm2_log.fittedvalues)
print('`modelo_sm2_log` R-squared:', tips[['pred_tip_log', 'tip']].corr().iloc[0,1]**2)

tips['pred_tip_pow'] = np.exp(modelo_sm2_pow.fittedvalues)
print('`modelo_sm2_pow` R-squared:', tips[['pred_tip_pow', 'tip']].corr().iloc[0,1]**2)

print('O modelo modelo_sm2_log obteve o resultado inferior, com', round(100*tips[['pred_tip_pow', 'tip']].corr().iloc[0,1]**2,2),'% de coeficiente de determinação em comparação com o modelo modelo_sm1_pow que obteve o resultado de',round(100*modelo_sm1_pow.rsquared,2),"%")
````

````python
filepath = 'https://raw.githubusercontent.com/Jefersonfranca/EBAC_Curso_Cientista_de_Dados/aacc1f2b11ef5a7d5492eac49f9f1de240cdac84/M%C3%B3dulo%2012%20Regress%C3%A3o%20I%20/database/previsao_de_renda.csv'

previsao_de_renda = pd.read_csv(filepath, index_col=0)
previsao_de_renda
````

````python
previsao_de_renda.dropna(inplace=True)
````

````python
y, X = patsy.dmatrices(formula_like='renda ~ tempo_emprego', data=previsao_de_renda)

reg = sm.OLS(y, X).fit()
reg.summary()
````

````python
previsao_de_renda['renda_predict'] = reg.fittedvalues

print('R-squared:', round(100*previsao_de_renda[['renda_predict','renda']].corr().iloc[0,1]**2,2), '%')
````

````python
plt.figure(figsize=(14,7))

sns.regplot(x='tempo_emprego', y='renda', data=previsao_de_renda, label='renda')
plt.plot(previsao_de_renda['tempo_emprego'], previsao_de_renda['renda_predict'], 'r.', label='renda_predict')

plt.legend()

plt.xticks(ticks=np.arange(stop=previsao_de_renda['tempo_emprego'].max(), step=1))
plt.yticks(ticks=np.arange(stop=previsao_de_renda['renda'].max(), step=25000))

plt.show()
````

````python
# 2) Análise de resíduos

plt.figure(figsize=(14,7))

previsao_de_renda['res'] = reg.resid

sns.scatterplot(x='tempo_emprego', y='res', data=previsao_de_renda)
plt.axhline(y=0, color='r', linestyle='--')

plt.xticks(ticks=np.arange(stop=previsao_de_renda['tempo_emprego'].max(), step=1))

plt.show()
````

````python
reg_log = smf.ols(formula='np.log(renda) ~ np.log(tempo_emprego)', data=previsao_de_renda).fit()

reg_log.summary()
````

````python
previsao_de_renda['renda_predict_log'] = np.exp(reg_log.predict(previsao_de_renda))

print('R-squared:', round(100*previsao_de_renda[['renda_predict_log', 'renda']].corr().iloc[0,1]**2, 2), '%')
````

````python
plt.figure(figsize=(14,7))

sns.regplot(x='tempo_emprego', y='renda', data=previsao_de_renda, label='renda')
plt.plot(previsao_de_renda['tempo_emprego'], previsao_de_renda['renda_predict_log'], 'r.', label='renda_predict_log')

plt.legend()

plt.xticks(ticks=np.arange(stop=previsao_de_renda['tempo_emprego'].max(), step=1))

plt.show()
````

````python
# 
plt.figure(figsize=(14,7))

previsao_de_renda['res'] = reg_log.resid

sns.scatterplot(x='tempo_emprego', y='res', data=previsao_de_renda)
plt.axhline(y=0, color='r', linestyle='--')

plt.xticks(ticks=np.arange(stop=previsao_de_renda['tempo_emprego'].max(), step=1))

plt.show()
````

````python
# 3) Ajuste um modelo de regressão linear múltipla para explicar renda (ou uma transformação de renda) de acordo com as demais variáveis

y, X = patsy.dmatrices(formula_like='renda ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda' 
                       '+ educacao + estado_civil + tipo_residencia + idade + tempo_emprego + qt_pessoas_residencia', data=previsao_de_renda)

reg_mult = sm.OLS(y, X).fit()
reg_mult.summary()
````

````python
# 4) Removendo as variáveis não significantes e ajustando o modelo.
y, X = patsy.dmatrices(formula_like='np.log(renda) ~ sexo + posse_de_imovel + idade + tempo_emprego', data=previsao_de_renda)

reg_mult = sm.OLS(y, X).fit()
reg_mult.summary()
````

````python
# Análisando os resíduos.

previsao_de_renda['res_log'] = reg_mult.resid

plt.figure(figsize=(14,7))

sns.scatterplot(x='tempo_emprego', y='res_log', data=previsao_de_renda)
plt.axhline(y=0, color='r', linestyle='--')

plt.xticks(ticks=np.arange(stop=previsao_de_renda['tempo_emprego'].max(), step=1))

plt.show()
````

````python
print('`renda_predict` R-squared:', round(previsao_de_renda[['renda_predict', 'renda']].corr().iloc[0,1]**2 * 100, 2),'%')

print('`renda_predict_log` R-squared:', round(previsao_de_renda[['renda_predict_log', 'renda']].corr().iloc[0,1]**2 * 100, 2),'%')

plt.figure(figsize=(14,7))
plt.plot(previsao_de_renda['tempo_emprego'], previsao_de_renda['renda'], '.', label='renda')
plt.plot(previsao_de_renda['tempo_emprego'], previsao_de_renda['renda_predict'], 'y+', alpha=.3, label='renda_predict')
plt.plot(previsao_de_renda['tempo_emprego'], previsao_de_renda['renda_predict_log'], 'r+', alpha=.3, label='renda_predict_log')

plt.legend()

plt.xticks(ticks=np.arange(stop=previsao_de_renda['tempo_emprego'].max(), step=1))

plt.show()
````
