````python
from pandas_datareader import data as web

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta
````

````python
%%time

# Lista de ativos
ativos = ['BBAS3.SA', 'PETR4.SA', 'ABEV3.SA']

# Data de início
dt_ini = '2022-01-01'

# Obtendo os dados históricos dos ativos
dados = yf.download(ativos, start=dt_ini)

# Exibindo as primeiras linhas do dataframe
dados.head()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Calculando a média móvel dos últimos 90 dias para os valores de fechamento ajustados
media_movel_90d = dados['Adj Close'].rolling(window=90).mean()

# Exibindo as ultimas linhas do dataframe com as médias móveis
media_movel_90d.tail(33)
````

````python
# Calculando o desvio padrão móvel dos últimos 90 dias para os valores de fechamento ajustados
desvio_padrao_90d = dados['Adj Close'].rolling(window=90).std()

# Exibindo as últimas 90 linhas do dataframe com as médias móveis e desvios padrão
print("Médias Móveis dos últimos 90 dias:")
print(media_movel_90d.tail(33))

print("\nDesvios Padrão Móveis dos últimos 90 dias:")
print(desvio_padrao_90d.tail(33))
````

````python
# Exibindo as últimas 90 linhas do dataframe com as médias móveis e desvios padrão
print("Médias Móveis dos últimos 90 dias:")
print(media_movel_90d.tail(90))

print("\nDesvios Padrão Móveis dos últimos 90 dias:")
desvio_padrao_90d.tail(90)
````

````python
# Convertendo para DataFrame longo para uso com Plotly Express
media_movel_90d_long = media_movel_90d.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Media_Movel_90d')

# Criando o gráfico de linhas
fig = px.line(media_movel_90d_long, x='Date', y='Media_Movel_90d', color='Ativo', template='plotly_white',
              title='Média Móvel de 90 Dias dos Ativos',
              labels={'Media_Movel_90d': 'Média Móvel (90 dias)', 'Date': 'Data', 'Ativo': 'Ativo'})

# Exibindo o gráfico
fig.show()
````

````python
# Convertendo para DataFrame longo para uso com Plotly Express
desvio_padrao_90d_long = desvio_padrao_90d.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Desvio_Padrao_90d')

# Criando o gráfico de linhas
fig = px.line(desvio_padrao_90d_long, x='Date', y='Desvio_Padrao_90d', color='Ativo', template='plotly_white',
              title='Desvio Padrão em Janela Móvel de 90 Dias dos Ativos',
              labels={'Desvio_Padrao_90d': 'Desvio Padrão (90 dias)', 'Date': 'Data', 'Ativo': 'Ativo'})

# Exibindo o gráfico
fig.show()
````

````python

# Calculando a amplitude (diferença entre valor máximo e mínimo) em uma janela móvel de 90 dias
amplitude_90d = dados['Adj Close'].rolling(window=90).apply(lambda x: x.max() - x.min(), raw=True)

# Convertendo para DataFrame longo para uso com Plotly Express
desvio_padrao_90d_long = desvio_padrao_90d.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Desvio_Padrao_90d')
amplitude_90d_long = amplitude_90d.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Amplitude_90d')

# Criando o gráfico de linhas para o desvio padrão
fig_desvio_padrao = px.line(desvio_padrao_90d_long, x='Date', y='Desvio_Padrao_90d', color='Ativo', template='plotly_white',
                            title='Desvio Padrão em Janela Móvel de 90 Dias dos Ativos', 
                            labels={'Desvio_Padrao_90d': 'Desvio Padrão (90 dias)', 'Date': 'Data', 'Ativo': 'Ativo'})

# Criando o gráfico de linhas para a amplitude
fig_amplitude = px.line(amplitude_90d_long, x='Date', y='Amplitude_90d', color='Ativo', template='plotly_white',
                        title='Amplitude em Janela Móvel de 90 Dias dos Ativos',
                        labels={'Amplitude_90d': 'Amplitude (90 dias)', 'Date': 'Data', 'Ativo': 'Ativo'})

# Exibindo os gráficos
fig_desvio_padrao.show()
fig_amplitude.show()
````

**Comparação e Conclusões:**

**Conclusões Semelhantes sobre Volatilidade:**

- Ambos os gráficos (desvio padrão e amplitude) revelam a volatilidade dos ativos. Se ambos os gráficos mostram picos em determinadas datas, isso indica períodos de alta volatilidade.
- Se um ativo apresenta um desvio padrão e amplitude consistentemente altos ao longo do tempo, isso sugere que o ativo é mais volátil.
- Comparando os dois gráficos, você pode observar que os picos de volatilidade (desvios padrão mais altos) devem coincidir com os picos de amplitude. Isso confirma que ambas as métricas estão capturando a mesma característica subjacente dos ativos: sua volatilidade.
- Períodos de baixa volatilidade também serão evidentes em ambos os gráficos, com baixos desvios padrão e baixa amplitude.


#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Definindo o ativo e data de início
ativo = 'PETR4.SA'

# Obtendo a data de 60 dias atrás
dt_fim = datetime.now()
dt_ini = dt_fim - timedelta(days=60)

# Obtendo os dados históricos do ativo
dados = yf.download(ativo, start=dt_ini.strftime('%Y-%m-%d'), end=dt_fim.strftime('%Y-%m-%d'))
````

````python
# Criando o gráfico de candlestick
fig = graph_objects.Figure(data=[graph_objects.Candlestick(x=dados.index,
                                     open=dados['Open'],
                                     high=dados['High'],
                                     low=dados['Low'],
                                     close=dados['Close'])])

# Adicionando título e rótulos
fig.update_layout(title=f'Candlestick de {ativo} - Últimos 60 Dias',
                  xaxis_title='Data',
                  yaxis_title='Preço',
                  xaxis_rangeslider_visible=False, template='plotly_white')

# Exibindo o gráfico
fig.show()
````
