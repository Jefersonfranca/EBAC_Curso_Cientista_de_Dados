#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

# 1. Calculando a amplitude de variação dos valores dos ativos.

```python
# Criando um novo DataFrame para armazenar os resultados
dados_amplitude = pd.DataFrame()

# Iterando sobre cada ativo
for ativo in ativos:
    # Selecionando dados do ativo específico
    dados_ativo = dados.xs(ativo, level=1, axis=1)
    
    # Calculando a amplitude e adicionando como nova coluna
    dados_ativo['Amplitude'] = dados_ativo['High'] - dados_ativo['Low']
    
    # Adicionando os dados ao DataFrame resultante, preservando o nome do ativo
    dados_ativo.columns = pd.MultiIndex.from_product([[ativo], dados_ativo.columns])
    dados_amplitude = pd.concat([dados_amplitude, dados_ativo], axis=1)

# Exibindo as primeiras linhas do DataFrame com a nova coluna
dados_amplitude.head()
```

```python
# Convertendo o DataFrame multi-index para um DataFrame plano
dados_amplitude_plano = dados_amplitude.stack(level=0).reset_index()

# Selecionando apenas as colunas 'Date', 'level_1' (ativo) e 'Amplitude'
dados_amplitude_final = dados_amplitude_plano[['Date', 'level_1', 'Amplitude']]

# Renomeando as colunas para maior clareza
dados_amplitude_final.columns = ['Data', 'Ativo', 'Amplitude']

# Exibindo as primeiras linhas do DataFrame final
dados_amplitude_final.head(6)
```

# 2. Gráfico mostrando a variação da amplitude

```python

fig = px.line(dados_amplitude_final, x='Data', y='Amplitude', color='Ativo', 
              title='Variação da Amplitude dos Ativos ao Longo do Tempo', template='plotly_white')
fig.show()
```

# 3. O ativo apresentou maior volatilidade

```python
# Calcular a média da amplitude diária para cada ativo
media_amplitude_por_ativo = dados_amplitude_final.groupby('Ativo')['Amplitude'].mean()

# Exibir a média da amplitude para cada ativo
print(media_amplitude_por_ativo)

# Identificar o ativo com a maior média de amplitude
ativo_mais_volatil = media_amplitude_por_ativo.idxmax()
maior_media_amplitude = media_amplitude_por_ativo.max()

print(f"O ativo mais volátil é {ativo_mais_volatil} com uma média de amplitude de {maior_media_amplitude:.2f}")
```
