```python
from scipy.stats import norm

p = 1 - norm.cdf(15.2, 15, 0.1)
print(f'A probabilidade de que um pacote tenha mais que 15.2 kg  aproximadamente {round(p * 100, 2)} %')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

# Calculando as probabilidades para Y1 (Normal Padro)
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)

# Calculando as probabilidades para Y2 (t-Student com 5 graus de liberdade)
prob_Y2 = stats.t.cdf(Y, df=5, loc=0, scale=1)

# Imprimindo os resultados
print("Probabilidades para Y1 (Normal Padro):")
for i in range(len(Y)):
    print(f"P(Y1 < {Y[i]:.4f}) = {prob_Y1[i]:.4f}")

print("\nProbabilidades para Y2 (t-Student com 5 graus de liberdade):")
for i in range(len(Y)):
    print(f"P(Y2 < {Y[i]:.4f}) = {prob_Y2[i]:.4f}")
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
import matplotlib.pyplot as plt

# Plotando as probabilidades
plt.figure(figsize=(10, 6))

plt.plot(Y, prob_Y1, label='Y1 - Normal Padro', linestyle='-', color='blue')
plt.plot(Y, prob_Y2, label='Y2 - t-Student (df=5)', linestyle='--', color='green')

plt.title('Funo de Distribuio Acumulada (CDF)')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade acumulada')
plt.legend()

plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui

# Calculando as probabilidades para Y1 (Normal Padro)
percentis_Y1 = stats.norm.ppf(Y, loc=0, scale=1)

# Calculando os percentis para Y2 (t-student com 5 graus de liberdade)
percentis_Y2 = stats.t.ppf(Y, df=5, loc=0, scale=1)

# Imprimindo os resultados
print("Percentis para Y1 (Normal Padro):")
for i in range(len(Y)):
    print(f"Percentil {Y[i]:.2f}: {percentis_Y1[i]:.4f}")

print("\nPercentis para Y2 (t-Student com 5 graus de liberdade):")
for i in range(len(Y)):
    print(f"Percentil {Y[i]:.2f}: {percentis_Y2[i]:.4f}")
```

6. Faa um grfico com os dados do item anterior.

```python
# Plotando os percentis
plt.figure(figsize=(10, 6))

plt.plot(Y, percentis_Y1, label='Y1 - Normal Padro', marker='o', linestyle='-', color='blue')
plt.plot(Y, percentis_Y2, label='Y2 - t-Student (df=5)', marker='s', linestyle='--', color='green')

plt.title('Percentis para Variveis Aleatrias Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Percentil')
plt.legend()

plt.grid(True)
plt.xticks(Y)
plt.tight_layout()
plt.show()
```
